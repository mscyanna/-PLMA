```python
def calculate_bmi(height, weight):
    try:
        height = float(height) / 100  # 將身高從公分轉換為公尺
        weight = float(weight)
        bmi = weight / (height ** 2)
        return bmi
    except ValueError:
        raise ValueError("輸入內容錯誤")

def interpret_bmi(bmi):
    if bmi < 18.5:
        return "過輕"
    elif 18.5 <= bmi < 24:
        return "適中"
    elif 24 <= bmi < 27:
        return "過重"
    elif 27 <= bmi:
        return "肥胖"

if __name__ == "__main__":
    height = input("請輸入身高(公分): ")
    weight = input("請輸入體重(公斤): ")

    try:
        bmi = calculate_bmi(height, weight)
        print("您的BMI為: {:.2f}".format(bmi))
        print("您的體位為:", interpret_bmi(bmi))
    except ValueError as e:
        print(e)
```

```python
def draw_christmas_tree(height):
    if height < 7:
        print("請輸入至少7層以上的聖誕樹樹葉高度")
        return

    for i in range(1, height + 1):#樹葉
        print(" " * (height - i) + "*" * (2 * i - 1))

    for _ in range(5):#樹幹
        print(" " * (height - 2) + "***")

if __name__ == "__main__":
    tree_height = int(input("請輸入聖誕樹樹葉高度(至少7層以上): "))
    draw_christmas_tree(tree_height)
```
